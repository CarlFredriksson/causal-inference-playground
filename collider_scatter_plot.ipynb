```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.normal(0, 1, 1000)
Y = -0.5 * X + np.random.normal(0, 2, len(X))
Z = np.array([-2 * X + 1 * Y + np.random.normal(0, 1, len(X)) > 0]).astype(int).squeeze()
```

```python
reg = LinearRegression().fit(np.expand_dims(X, axis=1), np.expand_dims(Y, axis=1))
X_pred = np.expand_dims(np.arange(-4, 4), axis=1)
Y_pred = reg.predict(X_pred)
```

```python
plt.scatter(X, Y)
plt.plot(X_pred, Y_pred, color="red")
```

```python
reg = LinearRegression().fit(np.expand_dims(X[Z==0], axis=1), np.expand_dims(Y[Z==0], axis=1))
X_pred = np.expand_dims(np.arange(-4, 4), axis=1)
Y_pred = reg.predict(X_pred)
plt.scatter(X[Z==0], Y[Z==0])
plt.plot(X_pred, Y_pred, color="red")
```

```python
reg = LinearRegression().fit(np.expand_dims(X[Z==1], axis=1), np.expand_dims(Y[Z==1], axis=1))
X_pred = np.expand_dims(np.arange(-4, 4), axis=1)
Y_pred = reg.predict(X_pred)
plt.scatter(X[Z==1], Y[Z==1])
plt.plot(X_pred, Y_pred, color="red")
```

```python
df = pd.DataFrame(
    data={
        "X": X,
        "Y": Y,
        "Z": Z
    }
)
df
```

```python
fig = px.scatter_3d(df, x="X", y="Y", z="Z", color="Z")
fig.show()
```
